#Let us write a for loop to get the unique values in each column and the type of the column. This is to make sure that we can identify the columns that needs to be modified into various data types such as int32, float 64, dates etc

```python
df_column = {}
for i in df.columns:
    df_column[i] = [df[i].unique(),df[i].dtype]
```

```python
df_column
```

Observations : 
Based on the observation of Unique values, 
1. The column names 'day', 'month', 'year', 'Temperature',' RH',' Ws' needs to be converted into the int32. 
2. The column names 'Rain ','FFMC', 'DMC', 'DC','ISI', 'BUI','FWI' are of type Object and needs to be converted into the float 64. 
3. Two column names have spaces in the heading and these needs to be eleminated for the ease of using the column names. The names of the columns are 'Classes  ',' RH',' Ws','Rain '
4. Based on the columns, 'day','month', 'year' the date column is being taken. 

```python
for i in ['day','month','year','Temperature',' RH',' Ws']:
    try : 
        df[i] = df[i].astype('int32')
    except Exception as e: 
        print('Exception occured in %s and the exception is %s',i,e)
```

#to cross check whether the columns have been converted into the int32.

```python
df.dtypes
```

```python
for i in ('Rain ','FFMC', 'DMC', 'DC','ISI', 'BUI','FWI'):
    try : 
        df[i] = df[i].astype('float64')
    except Exception as e : 
        print("Error occured in converting %s into the float %s",i,e)
```
